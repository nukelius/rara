```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline    
```

```python
x = np.arange(-1,3.1,0.1)   # start,stop,step
a = (4/np.pi)*np.cos(np.pi*x/2)
b = (4/np.pi)*(np.cos(np.pi*x/2) - np.cos(3*np.pi*x/2)/3)
c = (4/np.pi)*(np.cos(np.pi*x/2) - np.cos(3*np.pi*x/2)/3
              + np.cos(5*np.pi*x/2)/5)
d = (4/np.pi)*(np.cos(np.pi*x/2) - np.cos(3*np.pi*x/2)/3
              + np.cos(5*np.pi*x/2)/5
              - np.cos(7*np.pi*x/2)/7)
```

```python
plt.plot(x,a,x,b,x,c,x,d)

plt.xlabel('-1 =< x =< 3')
plt.ylabel('Partial sum value')
plt.title('Plot of four successive fourier expansions')
plt.legend(['Partial sum (a)', 'Partial sum (b)', 
           'Partial sum (c)', 'Partial sum (d)']) 

plt.show()
```

The function below generates Fourier coefficients and evaluates them for a given range of x

```python
def F_n_x(n,x_start,x_stop,x_step):
    cos_coeff = lambda x:((-1)**(x-1))/(2*x-1)
    x = np.arange(x_start,x_stop,x_step)
    return (np.array(
        [(4/np.pi)*
         (cos_coeff(n)*np.cos(0.5*(2*n-1)*np.pi*x))
             for n in np.arange(1,n+1,1)]
    )).sum(axis=0)
```

```python
[F_n_x(100,x,x+1,1) for x in [0,0.5,0.9,0.99,1.1,2]]
```

This is inline with what we would expect, because from -1 to 1 we would expect the value to be very close to 1, and 1 to 3, very close to -1. 


```python
f_n_99 = np.array([F_n_x(n,0.99,1,1) for n in np.arange(100,2100,100)]).flatten()
```

```python
f_n_999 = np.array([F_n_x(n,0.999,1,1) for n in np.arange(100,2100,100)]).flatten()
```
